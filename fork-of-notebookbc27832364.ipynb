```python
# import csv
import time

def coding_exercise():

    while True:
        name = input("Please enter your name: ").strip()
        if name:
            break
        else:
            print("Name cannot be empty. Please try again.")

    file_name = f"{name}_data.csv"
    
    with open(file_name, mode="w", newline='') as csvfile:
        fieldnames = ['Name', 'Timestamp', 'Time Delay (seconds)', 'Number Entered']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()

        last_entry_time = None

        print("\nEnter numbers between 0-9. Type 'quit' to exit at any time.")

        while True:
            user_input = input("Enter a number (0-9) or 'quit': ").strip().lower()

            if user_input == 'quit':
                print("Exiting program. Data saved to CSV.")
                break

            if user_input.isdigit():
                number = int(user_input)
                if 0 <= number <= 9:
                    current_time = time.time()
                    timestamp_readable = time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(current_time))

                    time_delay = 0.0
                    if last_entry_time is not None:
                        time_delay = current_time - last_entry_time

                    writer.writerow({
                        'Name': name,
                        'Timestamp': timestamp_readable,
                        'Time Delay (seconds)': f"{time_delay:.2f}",
                        'Number Entered': number
                    })
                    last_entry_time = current_time
                    print(f"Recorded: Name={name}, Number={number}, Timestamp={timestamp_readable}, Delay={time_delay:.2f}s")
                else:
                    print("Invalid input. Please enter a number between 0 and 9.")
            else:
                print("Invalid input. Please enter a number or 'quit'.")

if __name__ == "__main__":
    coding_exercise()
```
